```python
import forecastio
from datetime import datetime
from datetime import timedelta

api_key = "140fcefe299d96cb491074fb34a7c77d"
lat = 25.686613
lng = -100.316116
units_touse = "ca"

start_date = "2019-01-01"
stop_date  = "2019-12-31"

start = datetime.strptime(start_date, "%Y-%m-%d")
stop = datetime.strptime(stop_date, "%Y-%m-%d")

temps = dict()

while start <= stop:
    forecast = forecastio.load_forecast(api_key, lat, lng, time=start, units=units_touse)
    byHour = forecast.hourly()
    print ("reading day ",start)
    for hour in byHour.data:
        print (hour.time, hour.temperature)
        temps[hour.time] = hour.temperature
    start = start + timedelta(days=1)  # increase day one by one
```

```python
import pandas as pd
tempdf = pd.DataFrame(temps.items(), columns=['Date', 'Temperature'])
umbral = 7 # temp en Celisus, igual a 45ºF que se considera la condición para contar "chill hours"
tempdf[tempdf['Temperature'] < umbral].count()   #contar horas debajo de temperatura umbral
#tempdf.describe()
```
